```julia
function mp4_with_dynamic_label(index, xmin, xmax, ymin, ymax)
    fig = Figure(size = (400, 300))  # wider to make space for colorbar
    
    # Plot axis at (1,1)
    ax = Axis(fig[1, 1], aspect = DataAspect(), limits = (xmin, xmax, 0, 10))
    scatter!(ax, 10 .* rand(10), 10 .* rand(10))
    
    # Add heatmap
    xgrid = LinRange(xmin, xmax, 200)
    ygrid = LinRange(ymin, ymax, 200)
    heat_vals = [abs(sin((xmax - x)*(ymax - y))) for x in xgrid, y in ygrid]
    hm = heatmap!(ax, xgrid, ygrid, heat_vals, colormap = :plasma)
    
    # Add colorbar at (1,2)
    Colorbar(fig[1, 2], hm; label = "cue", width = 12)
    
    # Observable LaTeX label
  
    t_label = Observable(latexstring("t = $(index[1])"))
    text!(ax, (xmin + xmax)/2, ymax - 0.2 * (ymax - ymin), text = t_label, overdraw = true, fontsize = 40, color = :white)
    
    filename = tempname() * ".mp4"
    
    record(fig, filename, 1:length(index); framerate = 2) do i
        t_label[] = latexstring("t = $(index[i])")
    end
    
    return read(filename)
end
```

```julia
mp4_bytes = mp4_with_dynamic_label(collect(0:10), 0, 10, 0, 10);

b64 = base64encode(mp4_bytes)
width = 300
height = 300

html = """
<video width="$width" height="$height" controls autoplay loop muted>
    <source src="data:video/mp4;base64,$b64" type="video/mp4">
    Your browser does not support the video tag.
</video>
"""
display("text/html", html)
```
