```julia
using Pkg
using Plots
plotlyjs()
```

```julia
# Example 3D surface
x = 1:0.5:10
y = 1:0.5:10
z = [sin(xi) * cos(yi) for xi in x, yi in y]

Plots.plot(x, y, z, st = :surface, title = "3D Surface Plot",
     xlabel = "X", ylabel = "Y", zlabel = "Z")
```

```julia
using PlotlyJS


z = [sin(xi) + cos(yi) for xi in x, yi in y]

plt = Plot(
    surface(z = z),
    Layout(
        title = "Interactive 3D Surface",
        scene = attr(
            xaxis_title = "X",
            yaxis_title = "Y",
            zaxis_title = "Z"
        )
    )
)

display(plt)
```
