```julia
using Pkg
using FFMPEG
using Base64
using Plots
```

```julia
# 1. Create a simple animation and save as MP4
anim = @animate for i in 1:30
    plot(sin, 0, 2π, title="Frame $i", legend=false)
    scatter!([π*i/30], [sin(π*i/30)], color=:red)
end

mp4_path = "animation.mp4"
gif_path = "animation.gif"

# Save animation as MP4 using ffmpeg
mp4(anim, mp4_path; fps=15)

# 2. Function to display MP4 inline in Jupyter
function display_mp4(file_path)
    video_bytes = read(file_path)      #Reads the entire contents of the MP4 file at `file_path` into memory as a byte array named `video_bytes`
    b64 = base64encode(video_bytes)    #Converts the byte data into ASCII characters, making it safe to embed into HTML
    
    #Specify that we want to display the ASCII data as an MP4 video; also specify some other formatting stuff. 
    html = """                         
    <video width="480" controls>
      <source src="data:video/mp4;base64,$b64" type="video/mp4">
      Your browser does not support the video tag.
    </video>
    """
    display("text/html", html)
end

# Display the video
display_mp4(mp4_path)
```

```julia
using Plots
using Base64
using IJulia

# Step 1: Create animation and save GIF into an in-memory buffer
function create_gif_buffer()
    anim = Animation()
    for t in 1:30
        plt = plot(sin, 0, 2π, legend=false, title="Frame $t")
        scatter!([t*2π/30], [sin(t*2π/30)], color=:red)
        frame(anim, plt)
    end
    # Write the animation to an IOBuffer instead of disk
    io = IOBuffer()
    gif(anim, io, fps=15)
    seekstart(io)
    return take!(io)
end

# Step 2: Convert GIF bytes to MP4 bytes using ffmpeg via pipeline (requires ffmpeg installed)
function gif_to_mp4_bytes(gif_bytes)
    # ffmpeg command to read GIF from stdin and write mp4 to stdout, no file I/O on disk
    cmd = `ffmpeg -f gif -i pipe:0 -movflags +faststart -pix_fmt yuv420p -vf "scale=trunc(iw/2)*2:trunc(ih/2)*2" -f mp4 pipe:1`

    # Run ffmpeg as a pipeline, feed GIF bytes to stdin, capture mp4 bytes from stdout
    mp4_io = IOBuffer()
    open(cmd, "w", stdout=mp4_io) do ffmpeg_stdin
        write(ffmpeg_stdin, gif_bytes)
    end
    seekstart(mp4_io)
    return take!(mp4_io)
end

# Step 3: Display MP4 in Jupyter notebook cell
function display_mp4(mp4_bytes)
    b64 = base64encode(mp4_bytes)
    html = """
    <video controls autoplay loop muted style="max-width: 600px;">
      <source src="data:video/mp4;base64,$b64" type="video/mp4">
      Your browser does not support the video tag.
    </video>
    """
    display("text/html", html)
end

# Run everything
gif_bytes = create_gif_buffer()
mp4_bytes = gif_to_mp4_bytes(gif_bytes)
display_mp4(mp4_bytes)
```
